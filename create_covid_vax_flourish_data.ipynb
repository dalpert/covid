## Functions

```python
def create_date_subset(date, df):
    df_subset = df[df.Date == date]
    df_slim = df_subset[['Location', 'Administered_Dose1_Pop_Pct']]
    df_slim.columns = ['state', date]
    df_slim.sort_values(by='state', inplace=True)
    return(df_slim)
```

```python
create_date_subset(dates[5], covid_vax_data_filt)
```

```python
dates = list(covid_vax_data_filt.Date.unique())
dates = [d for d in dates if (('2021' in d) & (d >= '02/13/2021'))]
dates.sort()
```

```python
n = len(dates)
df = create_date_subset(dates[n-1], covid_vax_data_filt)
for date in dates[1:]:
    d_slice = create_date_subset(date, covid_vax_data_filt)
    df = df.merge(d_slice)
```
