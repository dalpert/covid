```python
import numpy as np
import pandas as pd
```

```python
county_results = pd.read_csv('president_county_candidate.csv')
county_results.head(10)
```

```python
county_results_2p = county_results[county_results['candidate'].isin(['Joe Biden', 'Donald Trump'])]
state_results = county_results_2p.groupby(['state', 'candidate']).sum().reset_index()
state_results
```

```python
two_rows = state_results[state_results.state == 'Alabama']
two_rows[two_rows.candidate == 'Joe Biden'].total_votes
```

```python
def get_biden_proportion(two_rows):
    return(two_rows[two_rows.candidate == 'Joe Biden'].total_votes * 1.0 / sum(two_rows.total_votes))
```

```python
get_biden_proportion(two_rows)
```

```python
biden_prop = state_results.groupby('state').apply(get_biden_proportion).reset_index()
biden_prop.drop(['level_1'], axis=1, inplace=True)
biden_prop['state_abbr'] = ['AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'DC', 'FL', 'GA',
       'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA',
       'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY',
       'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX',
       'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY']
biden_prop['biden_margin'] = [(prop - .5) * 2 for prop in biden_prop.total_votes]
biden_prop.columns = ['state', 'biden_pct', 'state_abbr', 'biden_margin']
```
